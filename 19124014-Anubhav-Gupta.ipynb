<b>Function 4 </b>:
Here we define a function with <b>O(log(n))</b> and we shall plot it's graph to expect a lograthmic growth <br>
The program works on divide and conquer algorithm the famous binary search<br>

the famous algorithm works on divide and conquer where we divide the array till we reach a part where the middle value is the value to be found <br>
the plot here shows a lot of peaks that distort the regular expected path of a lograthmic growth,these peaks might be due to encountering the worst cases 
Here I've given a array to the function of merge sort whose size grows linearly and entries of which are generated randomly

```python
import time
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import random
time_ =[]
value = []

def binary_search(arr, x): 
    low = 0
    high = len(arr) - 1
    mid = 0
  
    while low <= high: 
  
        mid = (high + low) // 2
  
        # Check if x is present at mid 
        if arr[mid] < x: 
            low = mid + 1
  
        # If x is greater, ignore left half 
        elif arr[mid] > x: 
            high = mid - 1
  
        # If x is smaller, ignore right half 
        else: 
            return mid 
  
    # If we reach here, then the element was not present 
    return -1
  
  
# Test array 
arr = []   
# Function call 
 

for x in range(0,1000000,10000):
    for y in range(0,x):
        arr.append(random.randint(0,1000))
    t0=time.time()
    result = binary_search(arr, x) 
    t1 = time.time() - t0
    value.append(x)
    time_.append(t1)
    
    
plt.plot(value,time_) 
plt.xlabel('x - axis -Input') 
plt.ylabel('y - axis - Time') 
plt.title('Order O(log(n)') 
plt.show() 



colors = np.random.rand(100)
# area = (30 * 2)**2  # 0 to 15 point radii
plt.scatter(value, time_, c=colors, alpha=0.5)
plt.show()
```
